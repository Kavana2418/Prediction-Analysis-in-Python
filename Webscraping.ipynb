```python
# Step 1: Import necessary libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd  # Import pandas to save data to Excel
```

```python
def scrape_gov_news(url):
    # Send a request to the website
    response = requests.get(url)
    
    if response.status_code != 200:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
        return None
    
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Extract the headline
    title_tag = soup.find('h1')  # <h1> tag is typically used for titles
    if title_tag:
        headline = title_tag.text.strip()
    else:
        headline = 'No headline found'
    
    # Extract the article content (usually <div> or <p> tags are used for content)
    content_div = soup.find('div', class_='govspeak')  # The class name for content div
    if content_div:
        paragraphs = content_div.find_all('p')
        content = "\n".join([para.text for para in paragraphs])  # Combine all paragraph text
    else:
        content = 'No content found'
    
    # Print the results (optional)
    print(f"Headline: {headline}\n")
    print(f"Content: {content}\n")  # Print the full content
    
    return {'Headline': headline, 'Content': content}

# Step 3: Call the function with the URL
url = "https://www.gov.uk/government/news/uk-pavilion-3rd-largest-country-pavilion-at-semicon-taiwan"
article_data = scrape_gov_news(url)

# Step 4: Save the scraped data to an Excel file
if article_data:
    df = pd.DataFrame([article_data])  # Convert the data into a DataFrame
    df.to_excel('D:\webscraping\scraped_news3.xlsx', index=False)  # Save the DataFrame to an Excel file
    print("News data saved to 'scraped_news3.xlsx'")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime

# URL of the page to scrape
url = "https://www.gov.uk/search/news-and-communications"

# Send a request to fetch the webpage content
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    print("Successfully fetched the page")
else:
    print(f"Failed to fetch the page. Status code: {response.status_code}")
    exit()

# Parse the page content
soup = BeautifulSoup(response.content, 'html.parser')
#print(soup)

# Find all the news article links and associated metadata
news_list = soup.find_all('ul', class_='gem-c-document-list govuk-!-margin-bottom-5') #here
#print(news_list)

# Initialize an empty list to store news data
news_data = []

# Iterate through each <ul> and extract articles
for ul in news_list:
    articles = ul.find_all('li')
    for article in articles:
        link_tag = article.find('a')
        if link_tag:
            title = link_tag.get_text(strip=True)
            href = link_tag.get('href', '')
            full_url = f"https://www.gov.uk{href}"  # Complete the relative link
            
            # Extract the date from the metadata (if available on the main page)
            date_tag = article.find('time')  # Assuming the publication date is within a <time> tag
            if date_tag:
                article_date = date_tag.get_text(strip=True)
            else:
                article_date = "Date not found"

            # Now, send a request to the article's URL and scrape its content
            article_response = requests.get(full_url)
            if article_response.status_code == 200:
                article_soup = BeautifulSoup(article_response.content, 'html.parser')
                
                # Extract the headline (you can refine this based on the article structure)
                article_headline = article_soup.find('h1').get_text(strip=True)
                
                # Extract the content (assuming the content is in <div> or <p> tags, modify as needed)
                content_div = article_soup.find('div', class_='govspeak')
                if content_div:
                    paragraphs = content_div.find_all('p')
                    article_content = "\n".join([para.get_text(strip=True) for para in paragraphs])
                else:
                    article_content = "Content not found."
                
                #extract the author
                #author_tag = article_soup.find('p', class_='metadata')
                author_tag = article_soup.find('dd', class_='gem-c-metadata__definition')
                if author_tag:
                    article_author = author_tag.get_text(strip=True)
                else:
                    article_author = "Author not found."
                
                
                # Append the title, link, date, headline, and content to the news_data list
                news_data.append({
                    "Title": title,
                    "Link": full_url,
                    "Date": article_date,
                    "Author":article_author,
                    "Headline": article_headline,
                    "Content": article_content
                })

# Print the results, including the date
for news in news_data:
    print(f"Title: {news['Title']}")
    print(f"Link: {news['Link']}")
    print(f"Date: {news['Date']}")
    print(f"Author: {news['Author']}")
    print(f"Headline: {news['Headline']}")
    print(f"Content: {news['Content']}\n")

# Optionally, save the data to a DataFrame and then to an Excel file
if news_data:
    df = pd.DataFrame(news_data)
    df.to_excel('D:\\webscraping\\news_data_with_content_and_date13.xlsx', index=False)
    print("News data saved to 'news_data_with_content_and_date13.xlsx'")
```
